```python
import bs4
from urllib.request import urlopen
import re

# k10_outstanding = dict()
k10_floating = []
# k10_name = dict()


def stock_info(stock_cd):

	url_float = 'http://companyinfo.stock.naver.com/v1/company/c1010001.aspx?cmp_cd=' + stock_cd

	source = urlopen(url_float).read()
	soup = bs4.BeautifulSoup(source, 'lxml')

	#	Xpath = //*[@id="cTB11"]/tbody/tr[7]/td

	data = soup.find(id='cTB11').find_all('tr')[6].td.text

	tmp = data
	tmp = tmp.replace('\r', '')
	tmp = tmp.replace('\n', '')
	tmp = tmp.replace('\t', '')

	tmp = re.split('/', tmp)

	tmp[0] = tmp[0].replace('주', '')
	tmp[0] = tmp[0].replace(',', '')
	tmp[0] = tmp[0].replace(' ', '')

	outstanding = int(tmp[0])

	tmp[1] = tmp[1].replace('%', '')
	tmp[1] = tmp[1].replace(' ', '')

	floating = float(tmp[1])

	#	//*[@id="pArea"]/div[1]/div/table/tbody/tr[1]/td/dl/dt[1]/span
	name = soup.find(id='pArea').find('div').find('div').find('tr').find('td').find('span').text

	# k10_outstanding[stock_cd] = outstanding
	k10_floating.append(floating)
	# k10_name[stock_cd] = name
```
